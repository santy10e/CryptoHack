```python
import random
import sympy

# Función para generar claves pública y privada
def generar_claves(tam_primo_bits=1024):
    # Generar dos números primos aleatorios p y q
    p = sympy.randprime(2**(tam_primo_bits-1), 2**tam_primo_bits)
    q = sympy.randprime(2**(tam_primo_bits-1), 2**tam_primo_bits)

    # Calcular el módulo n y la función totient de Euler (phi_n)
    n = p * q
    phi_n = (p-1) * (q-1)

    # Elegir un exponente público e
    e = 65537  # Valor común para e (puede ser modificado según sea necesario, es el estándar)

    # Asegurarse de que e y phi_n sean coprimos
    while sympy.gcd(e, phi_n) != 1:
        e = random.randint(2, phi_n - 1)

    # Calcular el exponente privado d
    d = sympy.mod_inverse(e, phi_n)

    # Devolver las claves pública y privada
    clave_publica = (n, e)
    clave_privada = (n, d)

    return clave_publica, clave_privada

# Función para cifrar un mensaje utilizando la clave pública
def cifrar(mensaje, clave_publica):
    n, e = clave_publica
    # Aplicar el cifrado RSA al mensaje utilizando la clave pública
    mensaje_cifrado = [pow(ord(char), e, n) for char in mensaje]
    return mensaje_cifrado

# Función para descifrar un mensaje cifrado utilizando la clave privada
def descifrar(mensaje_cifrado, clave_privada):
    n, d = clave_privada
    # Aplicar el descifrado RSA al mensaje cifrado utilizando la clave privada
    mensaje_descifrado = [chr(pow(char, d, n)) for char in mensaje_cifrado]
    return ''.join(mensaje_descifrado)

if __name__ == "__main__":
    # Ejemplo de uso del algoritmo

    # Mensaje original que se va a cifrar
    mensaje_original = "Hola, mundo!"

    # Generar claves pública y privada
    clave_publica, clave_privada = generar_claves()

    # Cifrar el mensaje original con la clave pública
    mensaje_cifrado = cifrar(mensaje_original, clave_publica)
    print("Mensaje cifrado:", mensaje_cifrado)

    # Descifrar el mensaje cifrado con la clave privada
    mensaje_descifrado = descifrar(mensaje_cifrado, clave_privada)
    print("Mensaje descifrado:", mensaje_descifrado)
```

```python
import random
import math

def es_primo(num):
    """Verifica si un número es primo."""
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

def generar_primo():
    """Genera un número primo aleatorio."""
    while True:
        num = random.randint(100, 1000)
        if es_primo(num):
            return num

def generar_claves():
    """Genera claves pública y privada para RSA."""
    # Paso 1 y 2: Elegir dos números primos
    p = generar_primo()
    q = generar_primo()

    # Paso 3: Calcular n y z
    n = p * q
    z = (p - 1) * (q - 1)

    # Paso 4: Elegir un número primo relativo a z (k)
    k = elegir_k(z)

    # Paso 5: Calcular el inverso multiplicativo de k (e) módulo z
    e = inverso_multiplicativo(k, z)

    # La clave pública es (n, k) y la clave privada es (n, e)
    clave_publica = (n, k)
    clave_privada = (n, e)

    return clave_publica, clave_privada

def elegir_k(z):
    """Elige un número primo relativo a z."""
    while True:
        k = random.randint(2, z - 1)
        if math.gcd(k, z) == 1:
            return k

def inverso_multiplicativo(a, m):
    """Calcula el inverso multiplicativo de a módulo m."""
    m0, x0, x1 = m, 0, 1
    while a > 1:
        q = a // m
        m, a = a % m, m
        x0, x1 = x1 - q * x0, x0
    return x1 + m0 if x1 < 0 else x1

def cifrar(mensaje, clave_publica):
    """Cifra un mensaje utilizando la clave pública."""
    n, k = clave_publica
    return pow(mensaje, k, n)

def descifrar(cifrado, clave_privada):
    """Descifra un mensaje cifrado utilizando la clave privada."""
    n, e = clave_privada
    mensaje_descifrado = pow(cifrado, e, n)
    return mensaje_descifrado

# Ejemplo de uso
mensaje_original = 42

# Generar claves
clave_publica, clave_privada = generar_claves()

# Cifrar el mensaje
mensaje_cifrado = cifrar(mensaje_original, clave_publica)

# Descifrar el mensaje
mensaje_descifrado = descifrar(mensaje_cifrado, clave_privada)

# Mostrar resultados
print("Mensaje original:", mensaje_original)
print("Clave pública:", clave_publica)
print("Clave privada:", clave_privada)
print("Mensaje cifrado:", mensaje_cifrado)
print("Mensaje descifrado:", mensaje_descifrado)
```

```python
import random
from sympy import isprime, mod_inverse

def generar_numero_primo(bits):
    # Genera un número primo aleatorio con la cantidad especificada de bits
    while True:
        numero_primo = random.getrandbits(bits)
        if isprime(numero_primo):
            return numero_primo

def generar_claves_rsa(bits):
    # Selección de números primos p y q
    p = generar_numero_primo(bits)
    q = generar_numero_primo(bits)

    # Cálculo de N y phi(N)
    N = p * q
    phi_N = (p - 1) * (q - 1)

    # Selección de la clave pública e
    e = random.randrange(2, phi_N)
    while mod_inverse(e, phi_N) is None:
        e = random.randrange(2, phi_N)

    # Cálculo de la clave privada d
    d = mod_inverse(e, phi_N)

    # Devolver claves pública y privada
    clave_publica = (N, e)
    clave_privada = (N, d)

    return clave_privada, clave_publica

# Ejemplo de uso
bits_clave = 256  # Ajusta según el nivel de seguridad deseado
clave_privada, clave_publica = generar_claves_rsa(bits_clave)

print("Clave privada:", clave_privada)
print("Clave pública:", clave_publica)
```
